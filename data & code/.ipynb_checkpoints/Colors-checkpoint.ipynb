```python
import numpy as np # работа с матрицами
import pandas as pd # работа с наборами данных
import matplotlib.pyplot as plt # построение графиков
import umap # алгоритм UMAP
from mpl_toolkits.mplot3d import Axes3D # построение 3D графиков
```

```python
# загрузка данных
colors = np.array(pd.read_excel('colors.xlsx'))
```

```python
# построение 3D графика
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')
plt.tick_params(labelsize=14)
ax.scatter(xs=colors[:, 0], ys=colors[:, 1], zs=colors[:, 2], c= np.array(colors) / 255, s=200, alpha=1.0)
```

```python
# списки с гиперпараметрами
mindists = [0.1, 0.3, 0.6, 1.0]
neighs = [2, 3, 8]
```

```python
# снижение размерности UMAP и визуализация
for i in range(len(mindists)):
    for j in range(len(neighs)):
        embedding = umap.UMAP(min_dist=mindists[i], n_neighbors=neighs[j], random_state=4).fit_transform(colors)
        plt.figure(figsize=(5.9,6))
        plt.scatter(*embedding.T, s=170, c=np.array(colors)/255)
        plt.axis('off')
        plt.show()
```
