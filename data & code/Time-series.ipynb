```python
import numpy as np # работа с матрицами
import pandas as pd # работа с наборами данных
import matplotlib.pyplot as plt # построение графиков
import umap # алгоритм UMAP
import seaborn as sns # продвинутое построение графиков
from sklearn.preprocessing import StandardScaler # стандартизация данных
from scipy.spatial.distance import cdist # расчет расстояний между объектами из выборки
```

Загрузка данных и объединение в один датасет:

```python
clothes = pd.read_csv('sales/clothes.csv')
clothes = np.array(clothes["value"])
```

```python
electronics = pd.read_csv("sales/electronics.csv")
electronics = np.array(electronics["value"])
```

```python
food = pd.read_csv("sales/food.csv")
food = np.array(food["value"])
```

```python
furniture = pd.read_csv("sales/furniture.csv")
furniture = np.array(furniture["value"])
```

```python
garden = pd.read_csv("sales/garden.csv")
garden = np.array(garden["value"])
```

```python
gasoline = pd.read_csv("sales/gasoline.csv")
gasoline = np.array(gasoline["value"])
```

```python
hobby = pd.read_csv("sales/hobby.csv")
hobby = np.array(hobby["value"])
```

```python
care = pd.read_csv("sales/personal care.csv")
care = np.array(care["value"])
```

```python
vehicle = pd.read_csv("sales/vehicle.csv")
vehicle = np.array(vehicle["value"])
```

```python
names = ['Одежда', 'Электроника', 'Еда', 'Мебель', 'Сад', 'Топливо', 'Хобби', 'Уход за собой', 'Транспорт']
```

```python
data = np.vstack((clothes.T, electronics.T, food.T, furniture.T, garden.T, gasoline.T, hobby.T, care.T, vehicle.T))
```

```python
# стандартизация данных
scaler = StandardScaler()
data_st = pd.DataFrame(scaler.fit_transform(data.T)).T
```

```python
# списки с гиперпараметрами
mindists = [0.1, 0.3, 0.6, 1.0]
neighs = [2, 3, 5, 7]
```

```python
# снижение размерности UMAP и визуализация
for i in range(len(mindists)):
    for j in range(len(neighs)):
        embedding = umap.UMAP(min_dist=mindists[i], n_neighbors=neighs[j], random_state=4).fit_transform(data_st)
        plt.figure(figsize=(5.9,6))
        plt.scatter(*embedding.T, s=170, c=[i for i in range(9)], cmap='tab20b')
        plt.axis('off')
        plt.savefig('C:/Users/eliza/Documents/GitHub/CW/times/min={},n={}.png'.format(str(mindists[i]).replace(".",","), neighs[j]), transparent=True, bbox_inches='tight', pad_inches=0)
        plt.show()
```

```python
# расчет корреляции между всеми временными рядами из выборки
corr = np.array(pd.DataFrame(data).T.corr())
```

```python
# создание heatmap для корреляции
plt.figure(figsize = (18,7))
sns.set(font_scale=1.35)
sns.heatmap(corr, vmin=0,
            xticklabels=names,
            yticklabels=names, annot=True, cmap='YlGnBu')
plt.savefig('C:/Users/eliza/Documents/GitHub/CW/times/corr.png', bbox_inches='tight', pad_inches=0)
```

```python
# снижение размерности UMAP для исследования зависимости между корреляцией и расстояниями UMAP при n_neighbors=2
emb2 = umap.UMAP(min_dist=1.0, n_neighbors=2, random_state=4).fit_transform(data_st)
```

```python
# создание списка с расстояниями между объектами
dist = cdist(emb2, emb2)
dist2 = []
for i in range(8):
    for j in range(i+1, 9):
        dist2.append(dist[i, j])
```

```python
# создание списка с коэффициентами корреляции между объектами
corrs = []
for i in range(8):
    for j in range(i+1, 9):
        corrs.append(corr[i, j])
```

```python
# построение графика зависимости корреляции и расстояний UMAP + линия тренда
plt.figure(figsize=(18, 7))
plt.xlabel('Расстояние в низкоразмерном пространстве', fontsize=20)
plt.ylabel('Коэффициент корреляции', fontsize=20)
plt.scatter(dist2, corrs, s=100, c=[100/255, 149/255, 237/255])
plt.tick_params(labelsize=14)
z = np.polyfit(dist2, corrs, 1)
p = np.poly1d(z)
plt.plot(dist2, p(dist2), c=[70/255, 130/255, 180/255], linewidth=2.5)
plt.savefig('C:/Users/eliza/Documents/GitHub/CW/times/chart.png', transparent=True, bbox_inches='tight', pad_inches=0)
```

```python
# снижение размерности UMAP для исследования зависимости между корреляцией и расстояниями UMAP при n_neighbors=7
emb7 = umap.UMAP(min_dist=1.0, n_neighbors=7, random_state=4).fit_transform(data_st)
```

```python
# создание списка с расстояниями между объектами
dist = cdist(emb7, emb7)
dist7 = []
for i in range(8):
    for j in range(i+1, 9):
        dist7.append(dist[i, j])
```
